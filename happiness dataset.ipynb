```python
import pandas as pd
import matplotlib.pyplot as plt

# Load and filter the data for Mongolia
data = pd.read_csv('World Happiness Report Data - Sheet1.csv')
mongolia_data = data[data['Country name'] == 'Mongolia']

# Display the data
print("Happiness Score Data for Mongolia:")
display(mongolia_data)  # For Jupyter notebooks

# Plot settings
plt.figure(figsize=(10, 6))
plt.plot(mongolia_data['year'], mongolia_data['Life Ladder'], marker='o', linestyle='-', color='skyblue', label='Happiness Score')

# Annotate highest and lowest points
max_value = mongolia_data['Life Ladder'].max()
min_value = mongolia_data['Life Ladder'].min()
max_year = mongolia_data.loc[mongolia_data['Life Ladder'] == max_value, 'year'].values[0]
min_year = mongolia_data.loc[mongolia_data['Life Ladder'] == min_value, 'year'].values[0]

plt.annotate(f'Max: {max_value}', xy=(max_year, max_value), xytext=(max_year, max_value + 0.5),
             arrowprops=dict(facecolor='black', arrowstyle='->'), fontsize=10)
plt.annotate(f'Min: {min_value}', xy=(min_year, min_value), xytext=(min_year, min_value - 0.5),
             arrowprops=dict(facecolor='black', arrowstyle='->'), fontsize=10)

# Plot enhancements
plt.xlabel('Year')
plt.ylabel('Happiness Score (Life Ladder)')
plt.title("Happiness Score Trend in Mongolia")
plt.grid(True)
plt.legend()
plt.tight_layout()

# Show and save the plot
plt.savefig('mongolia_happiness_trend.png')  # Save the plot as an image
plt.show()
```
